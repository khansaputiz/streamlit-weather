```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
weather_dataset = pd.read_csv('weather.csv')
```

```python
weather_dataset.head()
```

```python
weather_dataset.shape
```

```python
weather_dataset['outcome'].value_counts()
```

```python
X = weather_dataset.drop(columns='outcome', axis=1)
Y = weather_dataset['outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y = weather_dataset['outcome']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 1, random_state=5)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('akurasi data training adalah = ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('akurasi data testing adalah ', test_data_accuracy)
```

```python
def predict_outcome(temperature, humidity, rainfall, sunshine, wind_speed):
    if rainfall > 0:
        if rainfall > 10:
            return 'rainy'
        elif humidity > 80:
            return 'rainy'
        else:
            return 'cloudy'
    elif sunshine > 8:
        if wind_speed >= 4:
            return 'windy'
        else:
            return 'sunny'
    elif sunshine > 5:
        return 'sunny'
    elif humidity > 80:
        return 'rainy'
    else:
        return 'cloudy'

# Data input
data = {
    'temperatur_rata-rata': [25.8, 27.1, 25.5, 26.5, 27.2, 28.4, 28.4, 26.3, 28.1, 28.8, 27.6, 27.8, 27.9, 28.6, 28.4, 27.7, 26.9, 27.4, 27.1, 27.4, 27.6, 28.1, 27.6, 27.9, 26.3, 27.5, 26.7, 25.6, 26.4, 27.3, 26.9],
    'kelembapan_rata-rata': [87, 83, 84, 85, 79, 72, 73, 84, 77, 73, 76, 76, 76, 72, 73, 76, 82, 79, 82, 78, 75, 76, 77, 76, 81, 79, 86, 92, 86, 80, 84],
    'curah_hujan': [17, 50.9, 0, 4.2, 16.8, 0, 0, 0, 7.4, 0, 0, 0, 0, 0.4, 0, 0, 3, 25, 0.2, 0, 3.2, 3.6, 0, 0.5, 1.4, 0, 12, 10.4, 38.1, 25, 0.6],
    'sunshine': [2.2, 2.2, 1.2, 0, 4.4, 4, 8.1, 10.4, 4.5, 9.7, 4.9, 0.6, 8.9, 5.7, 8.2, 7.5, 3.3, 0.2, 2.7, 1.2, 5.4, 6.8, 6.4, 3.7, 7.7, 4, 2.7, 0, 0, 1, 6.2],
    'kecepatan_angin_rata-rata': [3, 3, 2, 2, 3, 4, 4, 3, 2, 3, 2, 2, 2, 2, 4, 3, 2, 2, 2, 2, 3, 4, 3, 3, 3, 2, 2, 1, 2, 3, 3],
    'outcome': ['rainy', 'rainy', 'rainy', 'rainy', 'rainy', 'cloudy', 'windy', 'sunny', 'cloudy', 'sunny', 'cloudy', 'cloudy', 'sunny', 'cloudy', 'windy', 'sunny', 'rainy', 'rainy', 'rainy', 'cloudy', 'cloudy', 'cloudy', 'sunny', 'cloudy', 'rainy', 'cloudy', 'rainy', 'rainy', 'rainy', 'rainy', 'rainy']
}

# Prediksi dan evaluasi
correct_predictions = 0
for i in range(len(data['temperatur_rata-rata'])):
    temperature = data['temperatur_rata-rata'][i]
    humidity = data['kelembapan_rata-rata'][i]
    rainfall = data['curah_hujan'][i]
    sunshine = data['sunshine'][i]
    wind_speed = data['kecepatan_angin_rata-rata'][i]
    actual_outcome = data['outcome'][i]
    
    predicted_outcome = predict_outcome(temperature, humidity, rainfall, sunshine, wind_speed)
    
    if predicted_outcome == actual_outcome:
        correct_predictions += 1
    
    print(f"Data: {i+1}")
    print(f"Temperature: {temperature}°C, Humidity: {humidity}%, Rainfall: {rainfall}mm, Sunshine: {sunshine}h, Wind Speed: {wind_speed}m/s")
    print(f"Actual Outcome: {actual_outcome}, Predicted Outcome: {predicted_outcome}")
    print()

accuracy = correct_predictions / len(data['temperatur_rata-rata'])
print(f"Accuracy: {accuracy * 100:.2f}%")
```
